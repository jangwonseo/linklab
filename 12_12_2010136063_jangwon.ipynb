```python
class Counter:
    def __init__(self,count,step=1):
        self.count=count
        self.step=step
    def __str__(self):
        return '[Count(step:%s)] %s' %(self.step,self.count)
    def incr(self):
        value=self.count+self.step
        self.count = value
    def __call__(self):
        value=self.count+self.step
        self.count = value
    def __coerce__(self, y):
        return self.count, y
    def __cmp__(self, other):
        return self.count,other
    
c=Counter(10)
d=Counter(10,2)
c.incr()
d.incr()
c()
d()
c=c+5
d=d-5
print c
print d
print
print c>10
print d>10
print c<10
print d<10
print c==17
print d!=9
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 9])
    print t
```
